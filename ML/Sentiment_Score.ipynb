```python
import pandas as pd
```

```python
mbti_df = pd.read_csv("Resources/mbti.csv")
```

```python
len(mbti_df)
```

```python
posts = mbti_df['posts']
post_num = []
for i in range(0,len(posts)):
    post=posts[i].count("|||") + 1
    post_num.append(post)
```

```python
len(post_num)
```

```python
score = mbti_df[['type','posts']]
score.head()
# score['posts'][0]
```

```python
score = pd.concat([pd.Series(row['type'], row['posts'].split('|||'))              
                    for _, row in score.iterrows()]).reset_index()
```

```python
len(score)
```

```python
score = score.rename(columns={
    "index":"posts"
})
score.columns.values[1] = "type"
```

```python
score
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Target String Setting
target_string = score['posts']
```

```python
len(target_string)
```

```python
scores = []
for i in target_string:
    compound = analyzer.polarity_scores(i)["compound"]
    scores.append(compound)
```

```python
len(scores)
```

```python
score['sentiment_score_compound']= scores
```

```python
import numpy as np
sum = 0
average = []
a=0
for i in post_num:
    for j in range(a,a+i):
        sum += scores[j]
        avg = sum/i
    average.append(avg)
    a=a+i
len(average)
```

```python
mbti_df['sentiment_score']=average
```
